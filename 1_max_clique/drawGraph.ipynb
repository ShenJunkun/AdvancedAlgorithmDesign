```python
import genGraph
import networkx as nx
import matplotlib.pyplot as plt
```

```python
V = 15
W = 5
g, clique = genGraph.create_graph_with_hidden_clique(V, W)
# print(clique)
pos = nx.spring_layout(g)


nx.draw(g, pos, with_labels=True, node_color='skyblue', font_weight='bold', node_size=500, font_size=12)
nx.draw_networkx_edges(g, pos, edgelist=clique, edge_color='red')
plt.title("Maximum Clique Graph")  # 设置标题
plt.show()  # 显示图形
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# 创建一个简单的图
G = nx.Graph()
G.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 5), (5, 1)])

# 定义一些边需要的颜色
red_edges = [(1, 2), (4, 5)]
black_edges = list(set(G.edges()) - set(red_edges))

# 创建一个图形对象
plt.figure(figsize=(6, 6))

# 绘制图形，指定特定边的颜色
nx.draw(G, with_labels=True, edgelist=black_edges, edge_color='black')
nx.draw(G, with_labels=True, edgelist=red_edges, edge_color='red')

plt.title('Graph with Red and Black Edges')
plt.show()
```
